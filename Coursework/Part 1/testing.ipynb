```python
import cv2
import matplotlib.pyplot as plt

data = x_train.iloc[1:9690]

print(x_train.iloc[:1])

# plt.hist(data, bins=20, edgecolor='k')

# plt.xlabel('Pixel Value')
# plt.ylabel('Frequency')
# plt.title('Image Histogram')
```

```python
image_data = np.array(data, dtype=np.uint8)

equalized_image = cv2.equalizeHist(image_data)

plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.hist(image_data.ravel(), bins=256, range=(0, 256), color='b', alpha=0.7)
plt.title('Original Histogram')

plt.subplot(122)
plt.hist(equalized_image.ravel(), bins=256, range=(0, 256), color='r', alpha=0.7)
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()
```

```python
def apply_clahe(row):
    row = np.array(row, dtype=np.uint8)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    equalized_row = clahe.apply(row)
    return equalized_row

x_copy = x_train.copy()

print(x_copy.shape)

df_equalized = x_copy.apply(apply_clahe, axis=1)
```

```python
sample_size = 5
sample_rows = x_copy.sample(sample_size)
sample_equalized_rows = sample_rows.apply(apply_clahe, axis=1)

plt.figure(figsize=(15, 5))

for i in range(sample_size):
    plt.subplot(2, sample_size, i + 1)
    original_image = np.array(sample_rows.iloc[i]).reshape(48, 48)
    plt.imshow(original_image, cmap='gray')
    plt.title(f'Original {i+1}')
    plt.axis('off')

    plt.subplot(2, sample_size, sample_size + i + 1)
    equalized_image = np.array(sample_equalized_rows.iloc[i]).reshape(48, 48)
    plt.imshow(equalized_image, cmap='gray')
    plt.title(f'Equalized {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
