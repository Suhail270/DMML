```python
def apply_clahe(row):
    row = np.array(row, dtype=np.uint8)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    equalized_row = clahe.apply(row)
    return equalized_row

def brighten(row, target_mean_brightness=128, max_brightness_adjustment=50):
    current_mean_brightness = np.mean(row)
    brightness_adjustment = target_mean_brightness - current_mean_brightness
    brightness_adjustment = np.clip(brightness_adjustment, -max_brightness_adjustment, max_brightness_adjustment)
    scaling_factor = (current_mean_brightness + brightness_adjustment) / current_mean_brightness
    adjusted_row = row * scaling_factor
    adjusted_row[adjusted_row < 0] = 0
    adjusted_row[adjusted_row > 255] = 255
    return adjusted_row


def apply_gamma_correction(row, gamma=1.8):
    input_shape = row.shape
    row = np.array(row, dtype=np.uint8)
    corrected_row = ((row / 255.0) ** (1 / gamma) * 255).astype(np.uint8)
    corrected_row = corrected_row.reshape(*input_shape) 
    return corrected_row
```

```python
import matplotlib.pyplot as plt

x_copy = x_train.copy()

sample_size = 5
sample_indices = x_copy.sample(sample_size).index


plt.figure(figsize=(15, 10))

for i, idx in enumerate(sample_indices):
    plt.subplot(4, sample_size, i + 1)
    original_image = x_copy.iloc[idx].values.reshape(48, 48)
    plt.imshow(original_image, cmap='gray')
    plt.title(f'Original {i+1}')
    plt.axis('off')

    brightened_image = brighten(x_copy.iloc[idx].values)
    plt.subplot(4, sample_size, sample_size + i + 1)
    brightened_image = brightened_image.reshape(48, 48)
    plt.imshow(brightened_image, cmap='gray')
    plt.title(f'Brightened {i+1}')
    plt.axis('off')

    gamma_image = apply_gamma_correction(brightened_image)
    plt.subplot(4, sample_size, 2*sample_size + i + 1)
    plt.imshow(gamma_image, cmap='gray')
    plt.title(f'Gamma Corrected {i+1}')
    plt.axis('off')

    equalized_image = apply_clahe(gamma_image)
    plt.subplot(4, sample_size, 3*sample_size + i + 1)
    plt.imshow(equalized_image, cmap='gray')
    plt.title(f'Equalized {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from skimage.metrics import structural_similarity as ssim
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import to_categorical

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    shear_range=0.2,
    horizontal_flip=False,
)

x_train_array = x_train.values.reshape(-1, 48, 48, 1)
y_train_array = y_train.values

augmented_x = []
augmented_y = []

for i in range(x_train_array.shape[0]):
    original_image = x_train_array[i].reshape(48, 48) 
    augmented_x.append(original_image)
    augmented_y.extend(y_train_array[i])

    for j in range(5):
        augmented_image = original_image.copy()

        while True:
            expanded_original_image = augmented_image[:, :, np.newaxis]
            augmented_image = datagen.random_transform(expanded_original_image)
            augmented_image = augmented_image[:, :, 0]

            if len(augmented_x) > 0:
                last_augmented_image = augmented_x[-1]
                ssim_value = ssim(augmented_image, last_augmented_image, data_range=1.0)
            else:
                ssim_value = 0 

            if ssim_value < 0.95:
                break
        
        augmented_x.append(augmented_image)
        augmented_y.extend(y_train_array[i])

augmented_x_array = np.array(augmented_x)
augmented_x_reshaped = augmented_x_array.reshape(-1, 48 * 48)
augmented_df = pd.DataFrame(augmented_x_reshaped)
augmented_df['Class'] = augmented_y

print(augmented_df.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

sample_size = 20

sample_rows_class_5 = augmented_df[augmented_df['Class'] == 5].sample(sample_size)

sample_images = sample_rows_class_5.iloc[:, :-1].values
sample_labels = sample_rows_class_5['Class'].values

def display_images(images, labels, columns=5):
    num_images = len(images)
    rows = (num_images // columns) + 1

    plt.figure(figsize=(15, 5))
    for i in range(num_images):
        plt.subplot(rows, columns, i + 1)
        image = images[i].reshape(48, 48)
        plt.imshow(image, cmap='gray')
        plt.title(f'Class: {labels[i]}')
        plt.axis('off')

display_images(sample_images, sample_labels, columns=sample_size)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

num_images = 5

def unencode_labels(one_hot_labels):
    unencoded_labels = np.argmax(one_hot_labels, axis=1)
    return unencoded_labels


fig, axes = plt.subplots(num_images, 2, figsize=(8, 2 * num_images))

for i in range(num_images):

    index = np.random.randint(0, len(x_train_array))

    original_image = x_train_array[index].reshape(48, 48)

    expanded_original_image = original_image[:, :, np.newaxis]
    augmented_image = datagen.random_transform(expanded_original_image)

    augmented_image = augmented_image[:, :, 0]
    axes[i, 0].imshow(original_image, cmap='gray')
    axes[i, 0].set_title(f'Original Label: {unencode_labels(y_train_array)[index]}')
    axes[i, 0].axis('off')

    axes[i, 1].imshow(augmented_image, cmap='gray')
    axes[i, 1].set_title(f'Augmented Label: {unencode_labels(y_train_array)[index]}')
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()
```
