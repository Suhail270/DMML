```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np

x_train_50_features = pd.read_csv('x_train_50_features.csv')
y_train_all = pd.read_csv('y_train_all.csv')
y_train_all = y_train_all.rename(columns={'0': 'label'})

merged_data = pd.concat([x_train_50_features, y_train_all], axis=1)
merged_data.to_csv('eg.csv',index=False)
unique_classes = merged_data['label'].unique()
selected_features_by_class = {}
components_list = [5, 10, 20]

for class_label in unique_classes:
    class_data = merged_data[merged_data['label'] == class_label]
    x_train_class = class_data.drop('label', axis=1)
    y_train_class = class_data['label']
    selected_features_by_class[f'class_{class_label}'] = {}

    for n_components in components_list:
        pca = PCA(n_components=n_components)
        x_train_pca_class = pca.fit_transform(x_train_class)
        correlations_class = pd.DataFrame(x_train_pca_class).apply(lambda col: col.corr(y_train_class))

        # selected_components_class = correlations_class[abs(correlations_class[0]) >= 0.2].copy()
        # original_features_class = pd.DataFrame(pca.components_, columns=x_train_class.columns)
        # features_for_selected_components_class = original_features_class.loc[selected_components_class.index]
        # selected_features_by_class[f'class_{class_label}'][f'n_components_{n_components}'] = features_for_selected_components_class

        # selected_components_mask = abs(correlations_class[0]) >= 0.2

        selected_components_mask = np.full(n_components, True)
        if selected_components_mask.any():
            selected_components_class = correlations_class[selected_components_mask].copy()

            original_features_class = pd.DataFrame(pca.components_, columns=x_train_class.columns)
            features_for_selected_components_class = original_features_class.loc[selected_components_class.index]
            selected_features_by_class[f'class_{class_label}'][f'n_components_{n_components}'] = features_for_selected_components_class


for class_label, features_dict in selected_features_by_class.items():
    print(f"Class {class_label}:")
    for n_components, features in features_dict.items():
        print(f"  Features for {n_components} components:")
        print(features)
        print()

print("\nSelected Features by Class:")
print(selected_features_by_class)
```
